## Instagram User Lifestyle and Engagement Analysis

## 1. Introduction

Social media platforms have become an important part of modern life, influencing communication, entertainment, and digital marketing. Among them, Instagram is one of the most widely used platforms, allowing users to share photos, videos, stories, and reels while interacting through likes, comments, and messages. Instagram is not only used for personal networking but also plays a major role in brand promotion, influencer marketing, and online advertising. Understanding how users behave on this platform is essential for improving content strategies, enhancing user experience, and increasing business performance.

This project analyzes the Instagram Users Lifestyle Dataset to study user demographics, usage behavior, content preferences, and engagement patterns. By applying exploratory data analysis and visualization techniques, the project aims to identify trends in user activity, interaction behavior, and advertisement response. The insights obtained from this analysis can help businesses and marketers identify high-value user segments, design effective content strategies, and make data-driven decisions to improve engagement and marketing outcomes.

## 4. Importing Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 5.  Data Overview

```python
df = pd.read_csv("instagram_users_lifestyle_100k.csv")
print(df.shape)
df.head()
```

## 6. Basic Data Information

```python
df.info()
df.describe()
```

## 7. Data Cleaning

```python
# Checking Missing Values
df.isnull().sum()
```

```python
# Removing Duplicate Rows

df = df.drop_duplicates()
```

## 8. Exploratory Data Analysis (EDA)

```python
# checking the exact column names
print(df.columns)
```

```python
# Clean column names (remove extra spaces if any)
df.columns = df.columns.str.strip()

# Convert important numeric columns safely
df['daily_active_minutes_instagram'] = pd.to_numeric(df['daily_active_minutes_instagram'], errors='coerce')
df['user_engagement_score'] = pd.to_numeric(df['user_engagement_score'], errors='coerce')
df['followers_count'] = pd.to_numeric(df['followers_count'], errors='coerce')
```

## 8.1 Age Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=30, kde=True)
plt.title("Age Distribution of Instagram Users")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

# Interpretation:
Most Instagram users belong to the young age group between 18 and 30 years. This shows that Instagram is more popular among youth and young adults compared to older age groups.

## 8.2 Gender-wise User Distribution

```python
plt.figure(figsize=(6,4))
df['gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Users")
plt.show()
```

# Interpretation:
The bar chart shows the distribution of male and female users on Instagram. Both genders actively use the platform, indicating that Instagram has a balanced user base across genders.

## 8.3 Daily Active Minutes by Age Group

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='gender', y='daily_active_minutes_instagram', data=df)
plt.title("Daily Active Minutes by Gender")
plt.xlabel("Gender")
plt.ylabel("Daily Active Minutes on Instagram")
plt.show()
```

# Interpretation:
Most Instagram users belong to the young age group between 18 and 30 years. This shows that Instagram is more popular among youth and young adults compared to older age groups.

```python
# Create age groups
df['age_group'] = pd.cut(df['age'], 
                         bins=[13,20,30,40,50,60], 
                         labels=['Teen','20s','30s','40s','50+'])

# Plot
plt.figure(figsize=(8,5))
sns.boxplot(x='age_group', y='daily_active_minutes_instagram', data=df)
plt.title("Daily Active Minutes by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Daily Active Minutes on Instagram")
plt.show()
```

## 8.4 Followers vs Engagement Score

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='followers_count', y='user_engagement_score', data=df)
plt.title("Followers vs Engagement Score")
plt.xlabel("Followers Count")
plt.ylabel("User Engagement Score")
plt.show()
```

# Interpretation:
There is a positive relationship between followers count and engagement score. Users with more followers generally have higher engagement, but some users with fewer followers also show high engagement, indicating that content quality is important.

## 8.5 Content Preference Distribution

```python
plt.figure(figsize=(7,5))
df['content_type_preference'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Content Preference of Users")
plt.ylabel("")
plt.show()
```

# Interpretation:
There is a positive relationship between followers count and engagement score. Users with more followers generally have higher engagement, but some users with fewer followers also show high engagement, indicating that content quality is important.

## 8.6 Premium vs Normal Users

```python
plt.figure(figsize=(6,4))
df['uses_premium_features'].value_counts().plot(kind='bar')
plt.title("Premium vs Normal Users")
plt.xlabel("User Type (0 = Normal, 1 = Premium)")
plt.ylabel("Count")
plt.show()
```

# Interpretation:
The graph shows that most users are normal users, while a smaller portion uses premium features. However, premium users represent a valuable segment for the platform.

## 8.7 Ad Clicks per Day Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['ads_clicked_per_day'], bins=30)
plt.title("Ad Clicks Per Day Distribution")
plt.xlabel("Ad Clicks Per Day")
plt.ylabel("Count")
plt.show()
```

# Interpretation:
Most users click on very few advertisements per day, while only a small group of users clicks on many ads. This shows that only a limited portion of users actively interact with advertisements.

## 8.8 Daily Active Minutes vs Engagement Score

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='daily_active_minutes_instagram', y='user_engagement_score', data=df)
plt.title("Daily Active Minutes vs Engagement Score")
plt.xlabel("Daily Active Minutes on Instagram")
plt.ylabel("User Engagement Score")
plt.show()
```

# Interpretation:
Users who spend more time on Instagram generally have higher engagement scores. This confirms that active users are more likely to interact with content.

## 8.9 Engagement Score Distribution

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df['user_engagement_score'], fill=True)
plt.title("Distribution of User Engagement Score")
plt.xlabel("User Engagement Score")
plt.ylabel("Density")
plt.show()
```

# Interpretation:
Most users have medium engagement scores, while only a few users have very high engagement. This shows that highly engaged users form a small but important segment.

## 8.10 Top 10 Countries by Average Engagement

```python
top_countries = df.groupby('country')['user_engagement_score'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_countries.plot(kind='bar')
plt.title("Top 10 Countries by Average Engagement Score")
plt.xlabel("Country")
plt.ylabel("Average Engagement Score")
plt.show()
```

# Interpretation:
Some countries show significantly higher average engagement scores. These countries are ideal targets for marketing campaigns and business promotions.

## 8.11 Correlation Heatmap

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm')
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

# Interpretation:
The heatmap shows strong positive correlations between daily active minutes, session length, likes, comments, and engagement score. This indicates that higher activity leads to higher engagement on Instagram.
